```python
# import required libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
# load the dataset
data = pd.read_csv("breast-cancer-wisconsin.csv",na_values='?')
data.dropna(inplace=True)
```

```python
# Split features and target variable
X = data.drop(columns=["Class", "Sample"])
y = data["Class"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=602)
```

```python
# Train the C5.0 model
c50_model = DecisionTreeClassifier()
c50_model.fit(X_train, y_train)
```

```python
# Evaluate the C5.0 model
c50_predictions = c50_model.predict(X_test)
print("Classification Report for C5.0:")
print(classification_report(y_test, c50_predictions))
# Calculate and print accuracy
accuracy = accuracy_score(y_test, c50_predictions)
print("Accuracy for C5.0:", accuracy)
```

```python
# Generate and print confusion matrix
conf_matrix = confusion_matrix(y_test, c50_predictions)
print("Confusion Matrix for C5.0:")
print(conf_matrix)
```

```python
# plot tree

plt.figure(figsize=(50,30), dpi=250)
plot_tree(c50_model,fontsize=20,filled=True,feature_names=X.columns.tolist());
```
