```python
# import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# import the dataset
data = pd.read_csv("wisc_bc_ContinuousVar.csv")
# check if any missing values
print(data[data.isna().any(axis=1)])
```

```python
# Split features and target variable
X = data.drop(['id', 'diagnosis'], axis=1)
y = data['diagnosis']
data.head(15)
```

```python
# scaling
scaler = StandardScaler()
Xscaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
Xscaled.head()
```

```python
# Spliting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=72)
```

```python
# Model Training
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Without using the standard scaler:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
```

```python
# USING THE STANDARD SCALER:

# Spliting the dataset with scaled attr
X_train, X_test, y_train, y_test = train_test_split(Xscaled, y, test_size=0.3, random_state=72)
```

```python
# Model Training
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy using the standard scaler:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
```
